```python

import numpy as np
from ImgConversionUtils import *
import cv2
from matplotlib import pyplot as plt
from skimage import draw,transform
import random


def GetOneImg():
    """ 返回一个Image对象 """
    imgs_dict_=Imgs_filepath_Dict()
    which_set=str()
    set_index=0
    random_i=random.randint(0,2)
    for k_ in imgs_dict_:
        if random_i==set_index:
            which_set=k_
            break
        set_index+=1
    img_root="/home/wzc/zlt_self/ZNewMyModel/data"
    rand_i=random.randint(0,imgs_dict_[which_set].__len__()-1)
    img_one=imgs_dict_[which_set][rand_i]
    img_filepath=os.path.join(img_root,(which_set+"/disc_small/image/"+img_one))
    print(f"Img_from--{which_set}---{img_one}")
    return Image.open(img_filepath)


def show_plt(imgs_tuple):
    """ 图像显示。传参：图像元组 返回值：无 """
    out_imgs=np.hstack(imgs_tuple)
    out_imgs = cv2.cvtColor(out_imgs,cv2.COLOR_RGBA2BGR)
    plt.imshow(out_imgs)
    

def HoughCircles_trans(img):
    """ 霍夫曼圆形拟合 """
    HoughCircles_img=img.copy()
    circle1 = cv2.HoughCircles(HoughCircles_img, cv2.HOUGH_GRADIENT, 1, 3000, param1=200, param2=0.9, minRadius=100, maxRadius=256)  #把半径范围缩小点，检测内圆，瞳孔
    # print(circle1.shape)
    circles = circle1[0, :, :]  # 提取为二维
    circles = np.uint16(np.around(circles))  # 四舍五入，取整
    # show_plt((circles,))
    return circles[0]

""" 执行--1、获得一张图片 2、通道分离 3、图像颜色处理 4、 """
""" 视盘拟合 """
im_random=GetOneImg()
(r_,g_,b_)=im_random.split()
img_r_aft=ImgEnhance_one(r_,Mode_=["color","bright","contrast"],Factor=[0.8,0.8,10])
""" PIL->CV  霍夫圆形拟合-->[圆心x，圆心y，半径]   r通道画图"""
img_r_aft_cv=np.asarray(img_r_aft)
info_circle=HoughCircles_trans(img_r_aft_cv)
xxx__=cv2.circle(img_r_aft_cv.copy(), (info_circle[0], info_circle[1]), int(info_circle[2]*1.3), 111, 7)  # 画圆
""" 彩色原图，画圆 """
im_random_cv=np.asarray(im_random)
xxx__raw=cv2.circle(im_random_cv.copy(), (info_circle[0], info_circle[1]), int(info_circle[2]*1.3), (5,193,96), 7)  # 画圆
```
